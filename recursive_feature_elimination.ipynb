```python
import pandas as pd
```

```python
# Read data.csv

data_df = pd.read_csv('data.csv')
data_df.head()
```

```python
data_df = data_df.dropna(subset=['birth_year'])
data_df.shape
```

```python
# birth_year is y to predict, person_id is id, and the rest are X. Run recursive_feature_elimination to find the best features to predict birth_year.

from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor

X = data_df.drop(columns=['birth_year', 'person_id'])

y = data_df['birth_year']

rfe = RFE(RandomForestRegressor(), n_features_to_select=5)
rfe.fit(X, y)

X.columns[rfe.support_]
```
